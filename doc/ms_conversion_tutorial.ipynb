```python
import time
from xradio.vis.convert_msv2_to_processing_set import convert_msv2_to_processing_set
import os

partition_scheme="ddi_intent_field"
#partition_scheme="ddi_state"

#storage_backend = "netcdf"
storage_backend = "zarr"

ms_path = "Antennae_North.cal.ms"
start = time.time()
os.system("rm -rf Antennae_North.cal.vis.zarr")
convert_msv2_to_processing_set(infile=ms_path,outfile="Antennae_North.cal.vis.zarr",partition_scheme=partition_scheme,parallel=True,storage_backend=storage_backend,overwrite=True)
print('Total time ',time.time()-start)
```

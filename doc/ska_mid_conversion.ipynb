## Download dataset

```python
from xradio.data.datasets import download
ms_file = "AA2-Mid-sim_00000.ms"
download(file=ms_file,source='dropbox')
```

## Convert to Processing Set

```python
from xradio.vis.convert_msv2_to_processing_set import convert_msv2_to_processing_set

partition_scheme="ddi_intent_field"
#partition_scheme="ddi_state"

chunks_on_disk = {'frequency':1,'time':20} #baseline, polarization
outfile = "AA2-Mid-sim_00000.vis.zarr"
convert_msv2_to_processing_set(infile=ms_file, outfile=outfile, 
                               partition_scheme=partition_scheme,
                               parallel=False,overwrite=True,
                               chunks_on_disk=chunks_on_disk)
```

## Processing Set

```python
ps_name = outfile

from xradio.vis.read_processing_set import read_processing_set
ps = read_processing_set(ps_name=outfile)
ps.summary()
```

# Multi MS v2 Conversion

Simple example how to convert multiple MSv2 into a single PS. This is achieved by setting  persistence_mode="a" in convert_msv2_to_processing_set.

```python
import toolviper

#toolviper.utils.data.download(file="Antennae_North.cal.lsrk.split.ms")

dir(toolviper.utils.data)

toolviper.utils.data.download(file="small_lofar.ms")
toolviper.utils.data.download(file="AA2-Mid-sim_00000.ms")
```

```python
from xradio.measurement_set import convert_msv2_to_processing_set

msv2_list = [
    "small_lofar.ms",
    "AA2-Mid-sim_00000.ms",
]   

outfile = "combined_lofar_aa2.ps.zarr"

for msv2 in msv2_list:
    convert_msv2_to_processing_set(
        in_file=msv2,
        out_file=outfile,
        parallel_mode="partition",
        persistence_mode="a", # Append to existing Processing Set
    )
```

```python
from xradio.measurement_set import open_processing_set
ps = open_processing_set(outfile)
ps.xr_ps.summary()
```
